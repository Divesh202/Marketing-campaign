##1.Handling outliers for Year_birth.

```python
# finding IQR, Q1, Q3, min for Year_Birth
q1 = df['Year_Birth'].quantile(0.25)
q3 = df['Year_Birth'].quantile(0.75)

iqr = q3 - q1
min_ = q1 - 1.5 * iqr

print('q1 = ',q1, '\n'
      'q3 = ',q3 ,'\n'
      'iqr = ',iqr, '\n'
      'minimum = ',min_)
```

```python
outlier_index = df[df['Year_Birth'] < 1940].index    # fetching the indexes of outlying rows.
```

```python
df.loc[outlier_index,'Year_Birth']=min_   # replacing the values with minimum value i.e 1932
```

```python
df.iloc[outlier_index]   # checking that it has been done.
```

# 3.Which marketing campaign was the most successful?

```python
camp = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']]
```

```python
dict = {}
dict['campaign'] = [i for i in camp.columns]
dict['total_customer'] = [camp[i].sum()  for i in camp.columns]

dict
```

```python
camp_result = pd.DataFrame(dict)
camp_result
```

```python
figure = plt.figure(figsize = (10,8))
ax = sns.barplot(data = camp_result , x = 'campaign' , y = 'total_customer')
plt.xticks(rotation = 45)
ax.bar_label(ax.containers[0]);
```

# Inference -
1. Last held campaign is the most successful campaign as we got `334` new customers.

# 5.Which products are performing best?

```python
prod = df[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']]
prod.head()
```

```python
dic1 = {}
dic1['product'] = [i for i in prod.columns]
dic1['total_earning'] = [prod[i].sum() for i in prod.columns]
```

```python
prod_df = pd.DataFrame(dic1)
prod_df
```

```python
figure = plt.figure(figsize = (10,5))
ax = sns.barplot(data = prod_df , x = 'product', y = 'total_earning')
plt.xticks(rotation = 45)
ax.bar_label(ax.containers[0]);
```

# Inference -
1. `Wines` are performing exceptionaaly well as it generates revenue of around `6.8 Lakhs`, which is followed by `Meat Products` generating revenue of `3.7 Lakhs`.
